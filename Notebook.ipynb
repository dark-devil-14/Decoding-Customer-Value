### 📅 Frequency Hierarchy (High to Low)

```python
# Applying the feature engineering to the data set

#Convert the object Male/Female into 1/0
df['Gender'] = df['Gender'].map({'Male':1, 'Female':0})

# Convert the object Subscription Status Yes/No into 1/0
df['Subscription Status'] = df['Subscription Status'].map({'Yes':1, 'No':0})

# Convert Discount Applied Yes/No into 1/0
df['Discount Applied'] = df['Discount Applied'].map({'Yes':1, 'No':0})

#Convert the Promo Code Used Yes/No into 1/0
df['Promo Code Used'] = df['Promo Code Used'].map({'Yes':1, 'No':0})

#Converting the Frequency of Purchases into numerical values 
# Mapping categorical frequency to ordinal integers
frequency_map = {
    'Weekly': 1, 
    'Bi-Weekly': 2, 
    'Fortnightly': 2, 
    'Monthly': 3, 
    'Every 3 Months': 4, 
    'Quarterly': 4, 
    'Annually': 5
}

df['Frequency of Purchases'] = df['Frequency of Purchases'].map(frequency_map)
df.head()
```

### 📊 Strategic Breakdown

| Score | Logic | Segment Name | Business Meaning |
| :--- | :--- | :--- | :--- |
| **2 + Low Spend** | Both used AND spend < Median | **High Risk** | Only shops when margins are razor-thin. High churn risk. |
| **2** | Both used | **Heavy Discounter** | Strong preference for stackable deals. |
| **1** | Either used | **Casual Saver** | Normal consumer behavior; uses available deals. |
| **0** | None used | **Full Price** | Highly loyal or convenience-driven; best margins. |

```python
# Discount dependency 

# Column which flags if a customer used BOTH a discount and a promo code
# 2 = Used Both, 1 = Used One, 0 = Used Neither
df['Discount_Dependency'] = (df['Discount Applied'] == 1).astype(int) + (df['Promo Code Used'] == 1).astype(int)

#High Rist Discount Flag
# Flagging users with low spend + discount usage
condition = [
    (df['Discount_Dependency'] == 2) & (df['Purchase Amount (USD)'] < df['Purchase Amount (USD)'].median()),    (df['Discount_Dependency'] == 2),
    (df['Discount_Dependency'] == 1),
    (df['Discount_Dependency'] == 0)
]

choices = [
    'High Risk (Bargain Hunter)', 
    'Heavy Discounter', 
    'Casual Saver', 
    'Full Price Shopper'
]

df['Discount_Risk_Profile'] = np.select(condition, choices, default='Other')
```

```python
# Estimate the Customer Lifetime Value (CLV)  (eCLV)
# Formula : eCLV = (Average Purchase Value) x (1 + Frequency of Previous Purchases)
df['eCLV'] = df['Purchase Amount (USD)'] * (1 + df['Frequency of Purchases'])
df.head()
```

```python
# Engagement Score the loyalty of the customer 
# Create a composite engagement score based on Subscription Status, Previous Purchase Frequency, and Review Rating, Frequency of Purchases and Discount Dependency

# Assigning weights to each component
# +3 Point for Subscription Status is 1 
# +3 Point for Previous Purchases greater than 15
# +2 Point for Review Rating greater than or equal to 4
# -3 Point for Discount Dependency of 2 (High Risk)
# -2 Point for Discount Dependency of 1 (Heavy Discounter)
# +2 Point for Frequency of Purchases between 1 and 2

def calculate_engagement_score(row):
    score = 0
    
    # 1. Subscription Loyalty (+2)
    if row['Subscription Status'] == 1:
        score += 3
        
    # 2. Purchase History (+2)
    if row['Previous Purchases'] > 15:
        score += 3

    # 3. High-Velocity Frequency (+1)
    # Target: Weekly (1) and Bi-Weekly (2) shoppers
    if 1 <= row['Frequency of Purchases'] <= 2:
        score += 2
        
    # 4. Customer Satisfaction (+2)
    # Using float conversion to ensure 'Unengaged' strings don't break the logic
    try:
        rating = float(row['Review Rating'])
        if rating >= 4:
            score += 2
    except (ValueError, TypeError):
        pass # Score remains 0 for non-numeric/missing ratings
        
    # 5. Discount Sensitivity (Penalties)
    if row['Discount_Dependency'] == 2:
        score -= 3
    elif row['Discount_Dependency'] == 1:
        score -= 2
        
    return score

# Applying the function to the dataframe
df['Engagement_Score'] = df.apply(calculate_engagement_score, axis=1)
```

### 📊 Engagement Score Logic

| Feature | Weight | Logic |
| :--- | :---: | :--- |
| **Subscription** | `+3` | Reward for recurring commitment. |
| **History** | `+3` | Reward for >15 previous purchases. |
| **Frequency** | `+2` | Reward for Weekly/Bi-Weekly shoppers. |
| **Satisfaction** | `+2` | Reward for Ratings ≥ 4. |
| **Discount Dep. (1)** | `-2` | Penalty for Casual Discount usage. |
| **Discount Dep. (2)** | `-3` | Penalty for "Bargain Hunter" behavior. |

**Final Scale:** $-3$ (Minimum) to $+10$ (Maximum)

```python
import scipy.stats as stats

# Example A: t-test (2 groups)
group_a = df[df["Discount_Risk_Profile"] == "Full Price Shopper"]["Purchase Amount (USD)"]
group_b = df[df["Discount_Risk_Profile"] == "Heavy Discounter"]["Purchase Amount (USD)"]

t_stat, p_val = stats.ttest_ind(group_a, group_b, equal_var=False)
print("t-test p-value:", p_val)

# Example B: ANOVA (3+ groups)
groups = [
    df[df["Discount_Risk_Profile"] == label]["Purchase Amount (USD)"]
    for label in df["Discount_Risk_Profile"].unique()
]

f_stat, p_val = stats.f_oneway(*groups)
print("ANOVA p-value:", p_val)

# Example C: Chi-square (categorical vs categorical)
contingency = pd.crosstab(df["Subscription Status"], df["Discount_Risk_Profile"])
chi2, p_val, dof, expected = stats.chi2_contingency(contingency)
print("Chi-square p-value:", p_val)
```
